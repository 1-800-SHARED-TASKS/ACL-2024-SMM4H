```python
import pandas as pd
df_gpt4    = pd.read_csv("/content/GPT4_Task5_015_Dev_preds (1).csv")
df_claude  = pd.read_csv("/content/ClaudeOpus_Task5_002_Dev_Preds (1).csv")
df_bart    = pd.read_csv("/content/bartL_task5_predictions_001_epoch_20 (1).csv")
df_deberta = pd.read_csv("/content/debertav3L_task5_predictions_001_epoch_20 (1).csv")
df_electra = pd.read_csv("/content/electraL_task5_predictions_001_epoch_20 (1).csv")
df_roberta = pd.read_csv("/content/robertaL_task5_predictions_001_epoch_20 (1).csv")
```

```python
df_gpt4 = df_gpt4[['tweet_id','label','pred']]
df_gpt4 = df_gpt4.rename(columns={'pred': 'gpt4'})
df_gpt4
```

```python
df_claude = df_claude[['tweet_id','label','pred']]
df_claude = df_claude.rename(columns={'pred': 'claude'})
df_claude
```

```python
df_bart = df_bart[['tweet_id','label','predicted_label']]
df_bart = df_bart.rename(columns={'predicted_label': 'bart'})
df_bart
```

```python
df_deberta = df_deberta[['tweet_id','label','predicted_label']]
df_deberta = df_deberta.rename(columns={'predicted_label': 'deberta'})
df_deberta
```

```python
df_roberta = df_roberta[['tweet_id','label','predicted_label']]
df_roberta = df_roberta.rename(columns={'predicted_label': 'roberta'})
df_roberta
```

```python
df_electra = df_electra[['tweet_id','label','predicted_label']]
df_electra = df_electra.rename(columns={'predicted_label': 'electra'})
df_electra
```

```python
df_gpt4.info()
df_claude.info()
df_bart.info()
df_deberta.info()
df_electra.info()
df_roberta.info()
```

```python
# Merging the dataframes on 'tweet_id' and keeping 'label' from only the first dataframe
merged_df = df_gpt4.merge(df_claude[['tweet_id', 'claude']], on='tweet_id') \
                   .merge(df_bart[['tweet_id', 'bart']], on='tweet_id') \
                   .merge(df_deberta[['tweet_id', 'deberta']], on='tweet_id') \
                   .merge(df_electra[['tweet_id', 'electra']], on='tweet_id') \
                   .merge(df_roberta[['tweet_id', 'roberta']], on='tweet_id')
```

```python
merged_df
```

```python
merged_df.info()
```

```python
prediction_columns = ['gpt4', 'claude', 'bart', 'deberta', 'electra', 'roberta']
merged_df['or'] = (merged_df[prediction_columns] == 1).any(axis=1).astype(int)
merged_df['maj_vote'] = (merged_df[prediction_columns].sum(axis=1) > len(prediction_columns) / 2).astype(int)
merged_df
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, precision_score, recall_score, f1_score
true_labels = merged_df['label']
predicted_labels = merged_df['maj_vote']  # or another predictions column based on your setup
cm = confusion_matrix(true_labels, predicted_labels)
precision = precision_score(true_labels, predicted_labels, pos_label=1)
recall = recall_score(true_labels, predicted_labels, pos_label=1)
f1 = f1_score(true_labels, predicted_labels, pos_label=1)
fig, ax = plt.subplots(figsize=(12, 12))
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap='Reds', ax=ax)
ax.set_title('Confusion Matrix')
plt.show()
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
```
