```python
import numpy as np
from sklearn.metrics import f1_score
def find_optimal_thresholds(df, labels, prob_columns):
    optimal_thresholds = {}
    for column in prob_columns:
        best_threshold = 0
        best_f1 = -1
        for threshold in np.linspace(0, 1, 100):
            predictions = (df[column] >= threshold).astype(int)
            f1 = f1_score(df[labels] == int(column[-1]), predictions)
            if f1 > best_f1:
                best_f1 = f1
                best_threshold = threshold
        optimal_thresholds[column] = best_threshold
        print(f"Best threshold for {column}: {best_threshold} with F1 score: {best_f1}")
    return optimal_thresholds
```

```python
prob_cols = ['prob_label_0', 'prob_label_1', 'prob_label_2', 'prob_label_3']
labels = 'Classification'
optimal_thresholds = find_optimal_thresholds(merged_df, labels, prob_cols)
optimal_thresholds
```

```python
import numpy as np
from itertools import product
from sklearn.metrics import f1_score
def calculate_f1_for_thresholds(df, prob_columns, labels, thresholds):
    predictions = np.zeros(len(df), dtype=int)
    for idx, column in enumerate(prob_columns):
        predictions[(df[column] >= thresholds[idx]).values] = idx
    f1 = f1_score(df[labels], predictions, average='macro')
    return f1
def optimize_thresholds(df, prob_columns, labels):
    best_thresholds = []
    best_f1 = 0
    threshold_options = np.linspace(0, 1, 20)  # Fewer points for faster computation
    for thresholds in product(threshold_options, repeat=len(prob_columns)):
        current_f1 = calculate_f1_for_thresholds(df, prob_columns, labels, thresholds)
        if current_f1 > best_f1:
            best_f1 = current_f1
            best_thresholds = thresholds
    print(f"Best thresholds: {best_thresholds} with Macro F1 Score: {best_f1}")
    return dict(zip(prob_columns, best_thresholds))
prob_cols = ['prob_label_0', 'prob_label_1', 'prob_label_2', 'prob_label_3']
labels = 'Classification'
optimal_thresholds = optimize_thresholds(merged_df, prob_cols, labels)
```
