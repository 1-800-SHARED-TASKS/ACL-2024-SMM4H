```python
import pandas as pd
df_test = pd.read_csv("/content/SMM4H_Task3_testposts_decoy_keywords (3).csv")
df_A2   = pd.read_csv("/content/BartLarge_task3_predictions_A2_epoch30.csv")
df_B2   = pd.read_csv("/content/BartLarge_task3_predictions_B2_epoch30.csv")
```

```python
df_test
```

```python
df_A2 = df_A2[['predicted_label']]
df_A2 = df_A2.rename(columns={'predicted_label': 'label_A2'})
df_A2
```

```python
df_B2 = df_B2[['predicted_label']]
df_B2 = df_B2.rename(columns={'predicted_label': 'label_B2'})
df_B2
```

```python
df_submission_2AB = pd.concat([df_test, df_A2, df_B2], axis=1)
df_submission_2AB
```

```python
df_submission_2AB['label_B2'] += 1
```
