```python
import pandas as pd
df_test = pd.read_csv("/content/SMM4H_Task3_testposts_decoy_keywords (3).csv")
df_A1   = pd.read_csv("/content/BartLarge_task3_predictions_A1_epoch30 (1).csv")
df_B1   = pd.read_csv("/content/BartLarge_task3_predictions_B1_epoch20.csv")
```

```python
df_test
```

```python
df_A1 = df_A1[['predicted_label']]
df_A1 = df_A1.rename(columns={'predicted_label': 'label_A1'})
df_A1
```

```python
df_B1 = df_B1[['predicted_label']]
df_B1 = df_B1.rename(columns={'predicted_label': 'label_B1'})
df_B1
```

```python
df_submission_1AB = pd.concat([df_test, df_A1, df_B1], axis=1)
df_submission_1AB
```
