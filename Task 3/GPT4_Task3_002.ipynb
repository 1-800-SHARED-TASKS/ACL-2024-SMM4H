```python
from tqdm import tqdm
```

```python
predictions_q1 = []
for index, row in tqdm(df_dev.iterrows(), total=df_dev.shape[0], desc="Classifying"):
      chat_completion = client.chat.completions.create(
          model="gpt-4-turbo-preview",
          messages=[
              {
                  "role": "system",
                  "content": "Did outdoor spaces and activities get mentioned? Respond only with a 1 for yes or 0 for no. Only one character (0/1) nothing else."
              },
              {
                  "role": "user",
                  "content": "Here is the text to classify: " + row['text']
              }
          ],
      )
      prediction_q1 = chat_completion.choices[0].message.content
      predictions_q1.append(prediction_q1)
df_dev['ans_q1'] = predictions_q1
df_dev
```

```python
predictions_q2 = []
for index, row in tqdm(df_dev.iterrows(), total=df_dev.shape[0], desc="Classifying"):
      chat_completion = client.chat.completions.create(
          model="gpt-4-turbo-preview",
          messages=[
              {
                  "role": "system",
                  "content": "Did outdoor spaces and activities have any impact on the user's mental health ? Respond only with a 1 for yes or 0 for no. Only one character (0/1) nothing else."
              },
              {
                  "role": "user",
                  "content": "Here is the text to classify: " + row['text']
              }
          ],
      )
      prediction_q2 = chat_completion.choices[0].message.content
      predictions_q2.append(prediction_q2)
df_dev['ans_q2'] = predictions_q2
df_dev
```

```python
predictions_q3 = []
for index, row in tqdm(df_dev.iterrows(), total=df_dev.shape[0], desc="Classifying"):
      chat_completion = client.chat.completions.create(
          model="gpt-4-turbo-preview",
          messages=[
              {
                  "role": "system",
                  "content": "Did outdoor spaces and activities have a positive impact on the user's mental health ? Respond only with a 1 for yes or 0 for no. Only one character (0/1) nothing else."
              },
              {
                  "role": "user",
                  "content": "Here is the text to classify: " + row['text']
              }
          ],
      )
      prediction_q3 = chat_completion.choices[0].message.content
      predictions_q3.append(prediction_q3)
df_dev['ans_q3'] = predictions_q3
df_dev
```

```python
columns_to_update = ['ans_q1', 'ans_q2', 'ans_q3']
for col in columns_to_update:
    df_dev[col] = df_dev[col].apply(lambda x: 0 if len(str(x)) > 1 else x)
```

```python
df_dev['ans_q1'] = df_dev['ans_q1'].astype(int)
df_dev['ans_q2'] = df_dev['ans_q2'].astype(int)
df_dev['ans_q3'] = df_dev['ans_q3'].astype(int)
```

```python
def calculate_ans(row):
    if row['ans_q1'] == 0:
        return 0
    if row['ans_q2'] == 0:
        return 2
    if row['ans_q3'] == 0:
        return 3
    return 1
df_dev['ans'] = df_dev.apply(calculate_ans, axis=1)
df_dev
```

```python
df_dev['Classification'] = df_dev['Classification'].astype(int)
```

```python
import pandas as pd
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
true_labels = df_dev['Classification']
predicted_labels = df_dev['ans']
cm = confusion_matrix(true_labels, predicted_labels, labels=pd.unique(true_labels))
plt.figure(figsize=(10, 7))
sns.heatmap(cm, annot=True, fmt="d", cmap="Reds",
            xticklabels=pd.unique(true_labels),
            yticklabels=pd.unique(true_labels))
plt.title('Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()
```

```python
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score
accuracy = accuracy_score(true_labels, predicted_labels)
f1 = f1_score(true_labels, predicted_labels, average='macro')  # Use 'weighted' for imbalanced classes
recall = recall_score(true_labels, predicted_labels, average='macro')
precision = precision_score(true_labels, predicted_labels, average='macro')
print(f"Accuracy: {accuracy}")
print(f"F1 Score: {f1}")
print(f"Recall: {recall}")
print(f"Precision: {precision}")
```
